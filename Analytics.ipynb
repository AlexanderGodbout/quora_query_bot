```python
import matplotlib.pyplot as plt
from matplotlib.dates import date2num
import matplotlib.dates as mdates


fig, ax = plt.subplots(1,1)
#scrape_dates = [date2num(reddit_scrape_dates), date2num(quora_scrape_dates)]
ax.hist(date2num(reddit_scrape_dates), alpha=0.5, bins = 50, cumulative=True, label='reddit')
ax.hist(date2num(quora_scrape_dates), alpha=0.5, bins = 50, cumulative=True, label='quora')

ax.xaxis.set_major_locator(mdates.AutoDateLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d.%m.%y'))
ax.legend(loc='best')
```
